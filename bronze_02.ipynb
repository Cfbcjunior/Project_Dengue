```python
##%pip install xlrd
from pyspark.sql.functions import col
##%pip install xlrd
##%restart_python
# Leitura dos arquivos CSV
df = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .csv("/Volumes/workspace/default/dengue/*.csv")

# Adiciona coluna com origem do arquivo
df = df.withColumn(
    "arquivo_origem",
    col("_metadata.file_path")
)

# Salva na camada Bronze
df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze_dengue")

print("Tabela bronze criada com sucesso!")
```
